```python
import pandas as pd
import re
```

```python
data = pd.read_csv('dog_breeds.csv')
```

```python
data.head()
```

```python
def extract_average(val):
    numbers = re.findall(r'\d+', val)
    numbers = list(map(int, numbers))
    if len(numbers) > 1:
        return sum(numbers) / len(numbers)
    elif numbers:
        return numbers[0]
    return None
```

```python
data.rename(columns={'Стоимость': 'Средняя стоимость'}, inplace=True)
```

```python
data['Средняя стоимость'] = data['Средняя стоимость'].apply(extract_average)
```

```python
data.rename(columns={'Длительность жизни': 'Средняя длительность жизни'}, inplace=True)
```

```python
data['Средняя длительность жизни'] = data['Средняя длительность жизни'].apply(extract_average)
```

```python
data.rename(columns={'Рождаемость': 'Средняя рождаемость'}, inplace=True)
```

```python
data['Средняя рождаемость'] = data['Средняя рождаемость'].apply(extract_average)
```

```python
data.rename(columns={'Рост в холке': 'Средний рост'}, inplace=True)
```

```python
data['Средний рост'] = data['Средний рост'].apply(extract_average)
```

```python
data.rename(columns={'Вес': 'Средний вес'}, inplace=True)
```

```python
data['Средний вес'] = data['Средний вес'].apply(extract_average)
```

```python
import pymorphy2
```

```python
with open('stopwords-ru.txt', 'r', encoding='utf-8') as f:
    stop_words = set(f.read().splitlines())
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
def lemmatize_text(text):
    words = text.split()
    lemmatized_words = [morph.parse(word)[0].normal_form for word in words if word.lower() not in stop_words]
    return ' '.join(lemmatized_words)
```

```python
data['Описание'] = data['Описание'].apply(lemmatize_text)
```

```python
data['Описание'][0]
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)
```

```python
data['Описание'] = data['Описание'].apply(remove_punctuation)
```
